While Loops involving Numbers
---
<a class="anchor" id="while"></a>

A while loop tests an initial condition. If that condition is true, the loop starts executing. Every time the loop finishes, the condition is reevaluated. As long as the condition remains true, the loop keeps executing. As soon as the condition becomes false, the loop stops executing. Here's an example:


```python
number = 5
while number < 10:
    print(number)
    number = number + 1
```

Notice that we could have also done this with a for loop:

```python
for i in range(5,10):
    print(i)
```

While loops are similar to for loops. While loops are typically better to use when you don't know how many iterations you'll need. The two major differences between for and while loops are that while loops need their counting variable  initialized in a separate line (number = 5) and then need to be manually updated (number = number + 1). 

Notice that your program won't work if you forgot to declare number = 5:

```python
while number < 10:
    print(number)
    number = number + 1
```

In addition, you will get an infinite loop if you don't include the "number = number + 1" line, because your number will always stay at 5 and thus always be less than 10:

```python
number = 5
while number < 10:
    print(number)
```

This error is called a Traceback error. 

We can also count backwards using while loops:

```python
number = 10
while number > 0:
    print(number)
    number = number -1
```

### While !=
We can use while loops to iterate until we have reached a certain value. Here's an example:

```python
number = 0
while number != 13:
    print(number,  'is not unlucky')
    number += 1
```

While Loops involving Strings
---
<a class="anchor" id="strings"></a>

We often use while loops to evaluate a user's input. Look at the following example:

```python
guess = input('Guess my password: ')
while guess != 'kanye':
    print('Incorrect, guess again!')
    guess = input('Guess my password: ')
print('You guessed correctly!')
```

We can also use while loops asking the user for more input and stopping only when the user presses enter. For example:

```python
friend = input('Give me a name of one of your friends: ')
while friend:
    print(friend, 'is one of your friends.')
    friend = input('Give me another name of your friends: ')
print('You have no more friends!')
```

While Loops involving Lists
---
<a class="anchor" id="lists"></a>

We'll often use while loops to add items to a list. Consider our previous friends example. If we want to store all of a user's friends in a list, we can type:


```python
friend_list = []
friend = input('Give me a name of one of your friends: ')
while friend:
    friend_list.append(friend)
    friend = input('Give me another name of your friends: ')

print('You have no more friends!')
print('Your friends are',friend_list )
```

To store the odd numbers between 2 and 22, we could type:

```python
odds = []
number = 3
while number < 22:
    odds.append(number)
    number = number + 2
print(odds)
```

Keeping Count
---
<a class="anchor" id="count"></a>

Suppose we wanted to keep track of how many reds are in a color list. We can declare a variable, red_count, to be zero, and update it every time we find a red color:

```python
colors = ['red', 'red', 'blue', 'red', 'yellow', 'blue', 'yellow', 'green']
red_count = 0
for color in colors:
    if color == 'red':
        red_count = red_count + 1
print(red_count)
```

There is a built-in Python function called count that performs the above calculation for us:

```python
colors = ['red', 'red', 'blue', 'red', 'yellow', 'blue', 'yellow', 'green']
print(colors.count('red'))
```

However, if we wanted to know how many unique colors were in the list, we could create a new list. See below.

```python
colors = ['red', 'red', 'blue', 'red', 'yellow', 'blue', 'yellow', 'green']
unique_list = []
for color in colors:
    if color not in unique_list:
        unique_list.append(color)
print('unique colors: ', len(unique_list))
```

Actually, as a short cut, we could turn the color list into a set.  We'll learn about sets later, but for now we can use the trick. A set only contains unique elements.

```python
colors = ['red', 'red', 'blue', 'red', 'yellow', 'blue', 'yellow', 'green']
unique_list = set(colors)
print('unique colors: ', len(unique_list))
```

Suppose we wanted to count how many times each color in the list occurred. We could type:

```python
colors = ['red', 'red', 'blue', 'red', 'yellow', 'blue', 'yellow', 'green']
unique_list = []
for color in colors:
    if color not in unique_list:
        unique_list.append(color)
        print(color, colors.count(color))
```

As another example, suppose you ask a group of students for their ages and then count how many of each age there are.

```python
age = input('What is your age? ')
age_list=[]

while age:
    age_list.append(age)
    age = input('What is your age? ')

unique_list = []
for age in age_list:
    if age not in unique_list:
        unique_list.append(age)
        print(age, age_list.count(age))    
```

Sorting Algorithms
---
<a class="anchor" id="sorting"></a>
Clearly, we can easily take two sorted lists, combine them, and sort them into one list:

```python
list1 = [1,3,5,7,9]
list2 = [0,2,4,6]
newlist = list1 + list2
newlist.sort()
print(newlist)
```

However, what if we wanted to create our own sorting algorithm? We would only want to go through each list once to maximize efficency. In this case, we would need at most m + n operations, if m was the size of list 1 and n was the size of list 2. Try to create your own sorting algorithm in the exercise below:

### Exercise - merge
Merge two lists that are already stored in alphabetical order in one list in alphabetical order. Your algorithm should work through the lists linearly, meaning it should only pass through each list once. For example, ["aa", "xx", "zz"] and ["bb", "cc"] should merge into ["aa", "bb", "cc", "xx", "zz"].

```python
def linear_merge(list1,list2):
    mergedlist=[]
    i=0
    j=0
    if len(list1)<=len(list2):
        minlength = len(list1)
        minlist = list1
        maxlist = list2
    else:
        minlength = len(list2)
        minlist = list2
        maxlist = list1
        
    while i < minlength:
        if minlist[i]<=maxlist[j]:
            mergedlist.append(minlist[i])
            i=i+1
        else: 
            mergedlist.append(maxlist[j])
            j=j+1
    mergedlist = mergedlist + maxlist[j:]

    return mergedlist


linear_merge(["aa", "xx", "zz"],["bb", "cc"])
#linear_merge([1,2,2], [-1,1.5,5])
```

Placeholder Printing
---
<a class="anchor" id="placeholder"></a>

Often, experience programmers will prefer to use placeholders for printing statements instead of commas or plus signs. Consider the following code that you are used to writing:

```python
word = 'kanye'
number = 32
boolean = True
print('My favorite word is ' + word + ' and my favorite number is ' + str(number) + ' and my favorite boolean is ' + str(boolean))
```

We can use placeholders to write this same sentence:

```python
word = 'kanye'
number = 32
boolean = True
print('My favorite word is %s and my favorite number is %s and my favorite boolean is %s' %(word, number, boolean))
```

Study the above code carefully. We put our sentence between quotation marks, we use placeholders %s wherever we want to input our variable, and we use %() at the end with whatever variables we need to insert inside. Here's another example:

```python
colors = ['red', 'red', 'blue', 'red', 'yellow', 'blue', 'yellow', 'green']
unique_list = []
for color in colors:
    if color not in unique_list:
        unique_list.append(color)
        print('Color: %s, Color Count: %s' %(color, colors.count(color)))
```

What does %s stand for? It represents a string. As long as you want to print out a number as a number in a sentence, you are fine to refer to it as a string. However, if we need it to be considered a number, use %d:

```python
number1 = 32
number2 = 5
print('%d + %d = %d' % (number1, number2, number1+number2))
```

In addition, we can use placeholders to format numbers nicely. %f refers to float (decimals). If you want to truncate a number to two decimal places, you could type:


```python
number = 5172.8397893
print('%0.2f'% (number))
```
